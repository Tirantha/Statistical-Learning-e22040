# Q. Statistical Mechanics of the Quantum Oscillator

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

where $\hbar$ is the reduced Planck constant and $k_B$ is the Boltzmann constant.

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.
* Use the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$ to prove that your expression for $C_V$ can be compactly rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$




5. **Asymptotic Behavior (Discussion):**
Briefly describe the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$). Does it match the classical equipartition theorem prediction?



### **1. The Partition Function**

The canonical partition function is given by the sum over all energy states:

$$Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$$

Let $\beta = \frac{1}{k_B T}$. Substitute the energy eigenvalues $E_j = \hbar \omega \left(j + \frac{1}{2}\right)$:

$$Z(T) = \sum_{j=0}^\infty e^{-\beta \hbar \omega \left(j + \frac{1}{2}\right)}$$

We can factor out the constant term (the zero-point energy contribution) from the summation:

$$Z(T) = e^{-\frac{\beta \hbar \omega}{2}} \sum_{j=0}^\infty e^{-\beta \hbar \omega j} = e^{-\frac{\beta \hbar \omega}{2}} \sum_{j=0}^\infty \left(e^{-\beta \hbar \omega}\right)^j$$

The summation is a standard infinite geometric series of the form $\sum_{j=0}^\infty r^j = \frac{1}{1 - r}$, where the common ratio is $r = e^{-\beta \hbar \omega}$. Since $\beta \hbar \omega > 0$, we know $|r| < 1$, guaranteeing convergence. 

Applying the geometric series formula:

$$Z(T) = e^{-\frac{\beta \hbar \omega}{2}} \left( \frac{1}{1 - e^{-\beta \hbar \omega}} \right)$$

Substituting $\beta = \frac{1}{k_B T}$ back into the equation yields the final simplified form:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$

---

### **2. Thermodynamic Potentials**

**Helmholtz Free Energy, $F(T)$:**
Using the definition $F(T) = -k_B T \ln Z$:

$$F(T) = -k_B T \ln\left( \frac{e^{-\beta \hbar \omega / 2}}{1 - e^{-\beta \hbar \omega}} \right)$$
$$F(T) = -k_B T \left[ -\frac{\beta \hbar \omega}{2} - \ln\left(1 - e^{-\beta \hbar \omega}\right) \right]$$

Since $k_B T \cdot \beta = 1$, this simplifies to:

$$F(T) = \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right)$$

**Internal Energy, $U(T)$:**
Using $U(T) = -\frac{\partial \ln Z}{\partial \beta}$:

$$\ln Z = -\frac{\beta \hbar \omega}{2} - \ln\left(1 - e^{-\beta \hbar \omega}\right)$$

Differentiating with respect to $\beta$:

$$-\frac{\partial \ln Z}{\partial \beta} = \frac{\hbar \omega}{2} + \frac{1}{1 - e^{-\beta \hbar \omega}} \cdot \frac{\partial}{\partial \beta}\left(1 - e^{-\beta \hbar \omega}\right)$$
$$U(T) = \frac{\hbar \omega}{2} + \frac{\hbar \omega e^{-\beta \hbar \omega}}{1 - e^{-\beta \hbar \omega}}$$

Divide the numerator and denominator of the second term by $e^{-\beta \hbar \omega}$:

$$U(T) = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

---

### **3. Entropy**

Using the relation $S = -\left(\frac{\partial F}{\partial T}\right)$, we differentiate the Helmholtz Free Energy:

$$S = -\frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) \right]$$

Applying the product rule and chain rule to the second term:

$$S = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) - k_B T \left[ \frac{1}{1 - e^{-\frac{\hbar \omega}{k_BT}}} \cdot \left(-e^{-\frac{\hbar \omega}{k_BT}}\right) \cdot \left(\frac{\hbar \omega}{k_B T^2}\right) \right]$$
$$S = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{T} \frac{e^{-\frac{\hbar \omega}{k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$

Multiply the numerator and denominator of the fraction by $e^{\frac{\hbar \omega}{k_BT}}$:

$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{1}{T} \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

**Verification of $U = F + TS$:**
Let's evaluate $F + TS$ using our derived expressions:

$$F + TS = \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\beta \hbar \omega}\right) \right] + T \left[ -k_B \ln\left(1 - e^{-\beta \hbar \omega}\right) + \frac{1}{T} \frac{\hbar \omega}{e^{\beta \hbar \omega} - 1} \right]$$

Distributing the $T$ in the entropy term cancels the logarithmic terms:

$$F + TS = \frac{\hbar \omega}{2} + k_B T \ln(\dots) - k_B T \ln(\dots) + \frac{\hbar \omega}{e^{\beta \hbar \omega} - 1}$$
$$F + TS = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

This precisely matches our derived expression for $U(T)$, successfully verifying the relation.

---

### **4. Heat Capacity and Hyperbolic Form**

The constant-volume heat capacity is the temperature derivative of internal energy, $C_V = \left(\frac{\partial U}{\partial T}\right)$. The zero-point energy ($\frac{\hbar \omega}{2}$) is constant, so its derivative is zero:

$$C_V = \frac{\partial}{\partial T} \left( \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1} \right)$$
$$C_V = \hbar \omega \cdot \left(-1\right)\left(e^{\frac{\hbar \omega}{k_BT}} - 1\right)^{-2} \cdot e^{\frac{\hbar \omega}{k_BT}} \cdot \left(-\frac{\hbar \omega}{k_B T^2}\right)$$
$$C_V = \frac{(\hbar \omega)^2}{k_B T^2} \frac{e^{\frac{\hbar \omega}{k_BT}}}{\left(e^{\frac{\hbar \omega}{k_BT}} - 1\right)^2}$$

**Converting to Hyperbolic Form:**
Let $x = \frac{\hbar \omega}{2k_BT}$. This implies $\frac{\hbar \omega}{k_BT} = 2x$. Substituting this into our $C_V$ expression:

$$C_V = k_B \left(\frac{\hbar \omega}{k_BT}\right)^2 \frac{e^{2x}}{(e^{2x} - 1)^2} = k_B (2x)^2 \frac{e^{2x}}{(e^x(e^x - e^{-x}))^2}$$

Factor the denominator:
$$C_V = 4 k_B x^2 \frac{e^{2x}}{e^{2x} (e^x - e^{-x})^2}$$

The $e^{2x}$ terms cancel out. Using the given identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$, we know $(e^x - e^{-x})^2 = (2 \sinh(x))^2 = 4 \sinh^2(x)$:

$$C_V = 4 k_B x^2 \frac{1}{4 \sinh^2(x)} = k_B \frac{x^2}{\sinh^2(x)}$$

Substituting $x$ back yields the required compact form:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$

---

### **5. Asymptotic Behavior (Discussion)**

In the high-temperature limit, $k_B T \gg \hbar \omega$. 
Defining our dimensionless variable $x = \frac{\hbar \omega}{2k_B T}$, this limit corresponds to $x \ll 1$ (as $T \to \infty$, $x \to 0$).

For very small values of $x$, we can use the Taylor series expansion for the hyperbolic sine function: $\sinh(x) \approx x + \frac{x^3}{3!} + \dots$ 
Truncating at the first order, $\sinh(x) \approx x$.

Substituting this approximation into our hyperbolic heat capacity equation:

$$C_V = k_B \frac{x^2}{\sinh^2(x)} \approx k_B \frac{x^2}{x^2} = k_B$$

**Does this match the classical equipartition theorem?**
**Yes.** According to the classical equipartition theorem, every independent quadratic degree of freedom contributes $\frac{1}{2} k_B$ to the heat capacity. A 1D harmonic oscillator has two such degrees of freedom: kinetic energy ($\propto p^2$) and potential energy ($\propto x^2$). 

Therefore, the classical prediction is $C_V = 2 \times \frac{1}{2} k_B = k_B$. 
The quantum harmonic oscillator perfectly recovers classical mechanics at high temperatures where the spacing between energy levels ($\hbar \omega$) becomes negligible compared to the thermal energy available ($k_B T$).

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$



### **1. The Objective and Constraints**

We want to find the probability density function $p(x)$ that maximizes the differential entropy:

$$H(X) = -\int_{-\infty}^\infty p(x) \ln p(x) dx$$

This maximization is subject to three constraints based on the properties of a valid probability distribution and the given moments:

1. **Normalization:** The total probability must equal 1.
   $$\int_{-\infty}^\infty p(x) dx = 1$$
2. **Mean:** The expected value is $\mu$.
   $$\int_{-\infty}^\infty x p(x) dx = \mu$$
3. **Variance:** The second central moment is $\sigma^2$.
   $$\int_{-\infty}^\infty (x-\mu)^2 p(x) dx = \sigma^2$$

To simplify the calculus, we can rewrite the second constraint as a central moment: $\int_{-\infty}^\infty (x-\mu) p(x) dx = 0$.

---

### **2. Formulating the Lagrangian**

We use the method of Lagrange multipliers to optimize the functional $H(X)$ subject to our integral constraints. Let $\lambda_0$, $\lambda_1$, and $\lambda_2$ be the Lagrange multipliers. The Lagrangian functional $\mathcal{L}$ is defined as:

$$\mathcal{L}[p] = \int_{-\infty}^\infty \left[ -p(x) \ln p(x) + \lambda_0 p(x) + \lambda_1 (x-\mu) p(x) + \lambda_2 (x-\mu)^2 p(x) \right] dx$$

To find the distribution that maximizes entropy, we take the functional derivative of $\mathcal{L}$ with respect to $p(x)$ and set it to zero:

$$\frac{\delta \mathcal{L}}{\delta p(x)} = -\ln p(x) - 1 + \lambda_0 + \lambda_1 (x-\mu) + \lambda_2 (x-\mu)^2 = 0$$

Solving for $p(x)$ yields the general form of the maximum entropy distribution:

$$p(x) = \exp\left( -1 + \lambda_0 + \lambda_1 (x-\mu) + \lambda_2 (x-\mu)^2 \right)$$

For mathematical convenience, we can group the constant terms into a new normalization constant $A = e^{\lambda_0 - 1}$, giving:

$$p(x) = A \exp\left( \lambda_1 (x-\mu) + \lambda_2 (x-\mu)^2 \right)$$

---

### **3. Solving for the Lagrange Multipliers**

We now apply our three constraints to determine the constants $A$, $\lambda_1$, and $\lambda_2$. To make the integration easier, let's use a change of variables: $y = x - \mu$. The limits of integration remain $(-\infty, \infty)$, and $dy = dx$. 

Our probability density function and constraints become:
$$p(y) = A \exp(\lambda_1 y + \lambda_2 y^2)$$
1. $\int_{-\infty}^\infty p(y) dy = 1$
2. $\int_{-\infty}^\infty y p(y) dy = 0$
3. $\int_{-\infty}^\infty y^2 p(y) dy = \sigma^2$

**Finding $\lambda_1$:**
For the density function to be a valid probability distribution, the integral must converge as $y \to \pm \infty$, which requires $\lambda_2 < 0$. 
Constraint 2 dictates that the first moment (the mean of $y$) is zero. For the integral of $y p(y)$ over a symmetric interval to be zero, $p(y)$ must be an even function. Therefore, the linear term in the exponent must vanish, meaning $\lambda_1 = 0$.

Our function simplifies to:
$$p(y) = A \exp(\lambda_2 y^2)$$

**Finding $\lambda_2$:**
We apply Constraint 3 (Variance). Using the standard Gaussian integral result $\int_{-\infty}^\infty y^2 e^{-a y^2} dy = \frac{1}{2a}\sqrt{\frac{\pi}{a}}$ (where $a = -\lambda_2$):

$$\sigma^2 = \int_{-\infty}^\infty y^2 A \exp(\lambda_2 y^2) dy = A \frac{1}{-2\lambda_2} \sqrt{\frac{\pi}{-\lambda_2}}$$

From Constraint 1 (Normalization), we know that $\int_{-\infty}^\infty A \exp(\lambda_2 y^2) dy = 1$, which gives $A \sqrt{\frac{\pi}{-\lambda_2}} = 1$. Substituting this into the variance equation:

$$\sigma^2 = (1) \cdot \frac{1}{-2\lambda_2}$$
$$\lambda_2 = -\frac{1}{2\sigma^2}$$

**Finding $A$:**
Substitute $\lambda_2$ back into the normalization equation to find $A$:

$$1 = A \sqrt{\frac{\pi}{1/(2\sigma^2)}} = A \sqrt{2\pi\sigma^2}$$
$$A = \frac{1}{\sqrt{2\pi\sigma^2}}$$

---

### **4. Final Distribution**

Substitute the evaluated constants $A$, $\lambda_1$, and $\lambda_2$ back into our general form for $p(x)$:

$$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( 0 - \frac{(x-\mu)^2}{2\sigma^2} \right)$$

This simplifies directly to the requested proof:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

Therefore, the continuous probability distribution over $(-\infty, \infty)$ that maximizes differential entropy for a specified mean and variance is the Gaussian (Normal) distribution.

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).






### **1. The System Constraint Equations**

Using the definition of expectation $\mathbb{E}[x_k \cdot \mathbb{I}(Y=1)] = \sum_{i} P(\mathbf{x}_i) \cdot x_{k,i} \cdot P(Y=1 \mid \mathbf{x}_i)$, we can translate the historical logs into a linear system of equations:

* **Constraint 1 (Financial Words):** $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = 0.05$
    $$0.40 p_A + 0.05 p_E = 0.05$$

* **Constraint 2 (Access Words):** $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = 0.15$
    $$0.25 p_B + 0.15 p_D + 0.05 p_E = 0.15$$

* **Constraint 3 (Bug Words):** $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = 0.16$
    $$0.05 p_C + 0.15 p_D = 0.16$$

We have **three equations** and **five unknowns** ($p_A, p_B, p_C, p_D, p_E$). Because $0 \le p_i \le 1$ for all probabilities, the solution space is tightly bounded, but not uniquely defined without an optimization objective.

### **2. Feasible Solution Bounds**

By analyzing the linear constraints, we can establish strict boundaries on how the AI must behave:
* From **Constraint 1**: The maximum possible value for $p_A$ occurs when $p_E = 0$, giving $p_A \le \frac{0.05}{0.40} = 0.125$. Even in the most extreme case, pure financial emails have at most a $12.5\%$ chance of routing to Technical.
* From **Constraint 3**: The maximum possible value for $p_D$ occurs when $p_C = 0$, giving $p_D \le \frac{0.16}{0.15} \approx 1.06$. Since probabilities cap at $1$, $p_D$ can theoretically be $1.0$. If $p_D = 1.0$, then $p_C = \frac{0.16 - 0.15}{0.05} = 0.20$.

### **3. The Maximum Entropy Formulation**

To find the single, least-biased probability distribution that satisfies these constraints, we maximize the conditional entropy $H(Y \mid X)$. For a discrete binary target, the MaxEnt solution takes the form of a logistic (sigmoid) function parameterized by Lagrange multipliers $\lambda_1, \lambda_2, \lambda_3$ corresponding to our three features:

$$P(Y=1 \mid \mathbf{x}) = \frac{\exp(\lambda_1 x_1 + \lambda_2 x_2 + \lambda_3 x_3)}{1 + \exp(\lambda_1 x_1 + \lambda_2 x_2 + \lambda_3 x_3)} = \sigma(\boldsymbol{\lambda}^T \mathbf{x})$$

By plugging our specific scenario vectors into this logistic model, we map the unknown probabilities to the Lagrange multipliers:

* $p_A = \sigma(\lambda_1)$
* $p_B = \sigma(\lambda_2)$
* $p_C = \sigma(\lambda_3)$
* $p_D = \sigma(\lambda_2 + \lambda_3)$
* $p_E = \sigma(\lambda_1 + \lambda_2)$

Substituting these back into our three constraint equations yields a non-linear system:

1.  $$0.40 \sigma(\lambda_1) + 0.05 \sigma(\lambda_1 + \lambda_2) = 0.05$$
2.  $$0.25 \sigma(\lambda_2) + 0.15 \sigma(\lambda_2 + \lambda_3) + 0.05 \sigma(\lambda_1 + \lambda_2) = 0.15$$
3.  $$0.05 \sigma(\lambda_3) + 0.15 \sigma(\lambda_2 + \lambda_3) = 0.16$$

This system can now be solved numerically (e.g., via gradient ascent or Newton-Raphson method) to find the unique values for $\lambda_1, \lambda_2,$ and $\lambda_3$.

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.





### **1. The Objective and Constraints**

We want to find the probability density function $p(S_T)$ over the continuous domain $(0, \infty)$ that maximizes the differential entropy:

$$H(S_T) = -\int_{0}^\infty p(S_T) \ln p(S_T) dS_T$$

This maximization is subject to four integral constraints based on the provided market data and the rules of probability:

1. **Normalization:** The total probability must equal 1.
   $$\int_{0}^\infty p(S_T) dS_T = 1$$
2. **Mean (Risk-Free Growth):**
   $$\int_{0}^\infty S_T p(S_T) dS_T = 102$$
3. **Variance/Volatility:**
   $$\int_{0}^\infty S_T^2 p(S_T) dS_T = 10600$$
4. **Logarithmic Skew (Downside Risk):**
   $$\int_{0}^\infty \ln(S_T) p(S_T) dS_T = 4.60$$

---

### **2. Formulating the Lagrangian**

We use the calculus of variations and introduce Lagrange multipliers ($\lambda_0, \lambda_1, \lambda_2, \lambda_3$) for each of our constraints. The Lagrangian functional $\mathcal{L}[p]$ is:

$$\mathcal{L}[p] = \int_{0}^\infty \left[ -p(S_T) \ln p(S_T) + \lambda_0 p(S_T) + \lambda_1 S_T p(S_T) + \lambda_2 S_T^2 p(S_T) + \lambda_3 \ln(S_T) p(S_T) \right] dS_T$$

To find the distribution that maximizes entropy, we take the functional derivative of $\mathcal{L}$ with respect to $p(S_T)$ and set it to zero:

$$\frac{\delta \mathcal{L}}{\delta p(S_T)} = -\ln p(S_T) - 1 + \lambda_0 + \lambda_1 S_T + \lambda_2 S_T^2 + \lambda_3 \ln(S_T) = 0$$

---

### **3. Solving for the Distribution**

Solving the variational equation for $p(S_T)$ yields:

$$\ln p(S_T) = (\lambda_0 - 1) + \lambda_1 S_T + \lambda_2 S_T^2 + \lambda_3 \ln(S_T)$$

Exponentiating both sides:

$$p(S_T) = \exp(\lambda_0 - 1) \cdot \exp(\lambda_1 S_T + \lambda_2 S_T^2 + \lambda_3 \ln(S_T))$$

We can simplify the logarithmic term using the property $\exp(\lambda_3 \ln(S_T)) = S_T^{\lambda_3}$. Furthermore, we can group the constant term $\exp(\lambda_0 - 1)$ into a single normalization constant, $Z$ (the partition function), which ensures the distribution integrates to 1. 

This results in the final algebraic form of the distribution:

$$p(S_T) = \frac{1}{Z} S_T^{\lambda_3} \exp\left(\lambda_1 S_T + \lambda_2 S_T^2\right)$$

---

### **4. Economic Intuition and Convergence Requirements**

The resulting MaxEnt distribution is a specialized 3-parameter exponential family distribution. It perfectly synthesizes the three market observations without assuming a rigid Gaussian shape:

* **The Power Law ($S_T^{\lambda_3}$):** Generated by the logarithmic constraint, this term controls the behavior of the distribution near zero. For the integral to converge at the lower bound ($S_T \to 0$), we must have $\lambda_3 > -1$. This heavily influences the "skew" and the probability of a market crash.
* **The Gaussian/Exponential Tail ($\exp(\lambda_1 S_T + \lambda_2 S_T^2)$):** Generated by the mean and variance constraints, this term controls the right tail (extreme upside). For the distribution to converge as $S_T \to \infty$, we must have $\lambda_2 < 0$. 

To finalize the exact numerical probability density function for Asset X, you would substitute this general form back into the constraint equations and solve numerically for $Z$, $\lambda_1$, $\lambda_2$, and $\lambda_3$ so that the distribution perfectly yields the observed values of $102$, $10,600$, and $4.60$.

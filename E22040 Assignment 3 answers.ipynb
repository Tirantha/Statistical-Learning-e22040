# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$


## Step 1: Joint random vector

Let

$$
\mathbf{Z} =
\begin{bmatrix}
\mathbf{X} \\
\mathbf{Y}
\end{bmatrix}
\in \mathbb{R}^{n+m}
$$

Assume:

$$
\mathbf{Z} \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})
$$

where

$$
\boldsymbol{\mu} =
\begin{bmatrix}
\boldsymbol{\mu}_X \\
\boldsymbol{\mu}_Y
\end{bmatrix}
$$

and

$$
\boldsymbol{\Sigma} =
\begin{bmatrix}
\boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\
\boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY}
\end{bmatrix}
$$

---

## Step 2: Joint density function

The joint density is:

$$
f_{X,Y}(\mathbf{x},\mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}}
\exp\left(
-\frac{1}{2}
\begin{bmatrix}
\mathbf{x}-\boldsymbol{\mu}_X \\
\mathbf{y}-\boldsymbol{\mu}_Y
\end{bmatrix}^T
\boldsymbol{\Sigma}^{-1}
\begin{bmatrix}
\mathbf{x}-\boldsymbol{\mu}_X \\
\mathbf{y}-\boldsymbol{\mu}_Y
\end{bmatrix}
\right)
$$

---

## Step 3: Definition of conditional density

By definition:

$$
f_{X|Y}(\mathbf{x}|\mathbf{y})
=
\frac{f_{X,Y}(\mathbf{x},\mathbf{y})}{f_Y(\mathbf{y})}
$$

So we must compute the marginal of $Y$ and simplify the exponent.

---

## Step 4: Key quadratic form decomposition

Using block matrix identities, the exponent can be rewritten as:

$$
(\mathbf{z}-\boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z}-\boldsymbol{\mu})
=
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\boldsymbol{\Sigma}_{X|Y}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*)
+
(\mathbf{y}-\boldsymbol{\mu}_Y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y)
$$

where:

---

### Conditional mean

$$
\boldsymbol{\mu}^*
=
\boldsymbol{\mu}_X
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y)
$$

---

### Conditional covariance (Schur complement)

$$
\boldsymbol{\Sigma}_{X|Y}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
$$

---

## Step 5: Marginal of $Y$

The marginal density of $\mathbf{Y}$ is:

$$
f_Y(\mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{y}-\boldsymbol{\mu}_Y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y)
\right)
$$

---

## Step 6: Form the conditional density

Now compute:

$$
f_{X|Y}(\mathbf{x}|\mathbf{y})
=
\frac{f_{X,Y}(\mathbf{x},\mathbf{y})}{f_Y(\mathbf{y})}
$$

### Cancellation step:

- All terms depending only on $\mathbf{y}$ cancel out
- Remaining terms depend only on $\mathbf{x}$ given $\mathbf{y}$

---

## Step 7: Final simplification

We obtain:

$$
f_{X|Y}(\mathbf{x}|\mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X|Y}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\boldsymbol{\Sigma}_{X|Y}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*)
\right)
$$

---

## Final Result

### Conditional distribution

$$
\mathbf{X}|\mathbf{Y}=\mathbf{y}
\sim
\mathcal{N}(\boldsymbol{\mu}^*, \boldsymbol{\Sigma}_{X|Y})
$$

---

### Conditional mean

$$
\boldsymbol{\mu}^*
=
\boldsymbol{\mu}_X
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y)
$$

---

### Conditional covariance

$$
\boldsymbol{\Sigma}_{X|Y}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
$$

# Q3) Continuous Fault Monitoring of a Spindle Motor

## 1. Joint Distribution of $(X, Y)$

Given:

$$
X \sim \mathcal{N}(0,1), \quad Q \sim \mathcal{N}(0,4), \quad Y = X + Q
$$

Since $X$ and $Q$ are independent and Gaussian:

$$
Y \sim \mathcal{N}(0, 1+4) = \mathcal{N}(0,5)
$$

---

### Mean vector

$$
\mu =
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

---

### Covariances

$$
\text{Var}(X)=1
$$

$$
\text{Var}(Y)=5
$$

$$
\text{Cov}(X,Y)=\text{Cov}(X, X+Q)
$$

$$
= \text{Var}(X) + \text{Cov}(X,Q)
$$

Since $X \perp Q$:

$$
\text{Cov}(X,Y)=1
$$

---

### Joint distribution

$$
\begin{bmatrix}
X \\
Y
\end{bmatrix}
\sim \mathcal{N}
\left(
\begin{bmatrix}
0\\
0
\end{bmatrix},
\begin{bmatrix}
1 & 1 \\
1 & 5
\end{bmatrix}
\right)
$$

## 2. Information Geometry (Why $X \notin \sigma(Y)$)

$Y$ is a **sum of two independent sources**:

- $X$ (temperature signal)
- $Q$ (mechanical noise)

So observing $Y=5$ does not uniquely determine $X$:

- High $Y$ could mean:
  - high temperature ($X$ large), or
  - high mechanical load ($Q$ large), or both

Thus:

$$
X \notin \sigma(Y)
$$

The system is **non-invertible due to noise mixing**.

---

## 3. Conditional Expectation $E[X|Y]$

For jointly Gaussian variables:

$$
E[X|Y=y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y-\mu_Y)
$$

Substitute values:

$$
E[X|Y=y] = 0 + \frac{1}{5}y
$$

---

### Final estimator

$$
E[X|Y] = \frac{1}{5}Y
$$

---

### Signal-to-noise interpretation

$$
\frac{\text{Cov}(X,Y)}{\text{Var}(Y)} = \frac{1}{5}
$$

So only **20% of the observed signal** is attributed to true temperature.

---

## 4. Residual Analysis

Let:

$$
\hat{X} = \frac{1}{5}Y
$$

Residual:

$$
Z = X - \hat{X}
$$

---

### (a) Mean of residual

$$
E[Z] = E[X] - E[\hat{X}] = 0 - \frac{1}{5}E[Y] = 0
$$

---

### (b) Orthogonality with $Y$

Compute:

$$
\text{Cov}(Z,Y) = \text{Cov}(X - \tfrac{1}{5}Y, Y)
$$

$$
= \text{Cov}(X,Y) - \tfrac{1}{5}\text{Var}(Y)
$$

$$
= 1 - \tfrac{1}{5}\cdot 5 = 0
$$

Thus:

$$
Z \perp Y \quad \text{(uncorrelated)}
$$

---

## 5. Differential Entropy

### (a) Entropy of $X$

For Gaussian:

$$
h(X) = \frac{1}{2}\log(2\pi e \sigma^2)
$$

$$
h(X) = \frac{1}{2}\log(2\pi e)
$$

---

### (b) Conditional variance

$$
\sigma_{X|Y}^2 = 1 - \frac{1^2}{5}

$$

---

### (c) Conditional entropy

$$
h(X|Y) = \frac{1}{2}\log(2\pi e \cdot \tfrac{4}{5})
$$

## 6. Mutual Information

$$
I(X;Y) = h(X) - h(X|Y)
$$

$$
I(X;Y)

### Numerical insight

### Percentage of uncertainty removed

\text{Cov}(X,Y)=1
$$

## 7. Diagnostic Interpretation

Claim: *"Best estimate is $Y$ itself."*

But:

$$
E[X|Y] = \frac{1}{5}Y \neq Y
$$

So $Y$ is a **biased estimator** of $X$:

$$
E[Y|X] = X + E[Q] = X
$$

but inversion is not valid due to noise.

---

### Why shrinkage happens

$$
E[X|Y] = \frac{\sigma_X^2}{\sigma_X^2 + \sigma_Q^2}Y
$$

So:

- if noise is large → coefficient small → heavy shrinkage
- estimator pulls values toward mean (0)

# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


# Probability in a Disk

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

---

## 1. Uniform with respect to area

A uniform distribution over the disk means probability is proportional to area.

Event:
$$
A = \{9 \le r \le 10\}
$$

Total area:
$$
A_{\text{total}} = \pi (10)^2 = 100\pi
$$

Inner disk area:
$$
A_{\text{inner}} = \pi (9)^2 = 81\pi
$$

Annulus area:
$$
A_A = 100\pi - 81\pi = 19\pi
$$

Probability:
$$
P(A) = \frac{19\pi}{100\pi} = \frac{19}{100} = 0.19
$$

---

## 2. Uniform radius and direction

Here:
- $r \sim \text{Uniform}[0,10]$
- $\theta \sim \text{Uniform}[0,2\pi)$

Event:
$$
A = \{9 \le r \le 10\}
$$

Probability:
$$
P(A) = \frac{10 - 9}{10} = \frac{1}{10} = 0.1
$$

---

## 3. Explanation

The two answers differ because the distributions are different.

In polar coordinates:
$$
dA = r \, dr \, d\theta
$$

So for a uniform area distribution, the density depends on $r$.  
A uniform radius distribution ignores this, giving too much weight to the center.

---

## Final Answers

$$
\boxed{P(A) = 0.19 \text{ (uniform area)}}
$$

$$
\boxed{P(A) = 0.10 \text{ (uniform radius)}}
$$

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

# Bertrand's Paradox: Random Chord Problem

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

---



For a circle of radius $R$, the side length of the inscribed equilateral triangle is:
$$
s = \sqrt{3}\,R
$$

A chord at distance $d$ from the center has length:
$$
L = 2\sqrt{R^2 - d^2}
$$

We want:
$$
L > \sqrt{3}R
$$

---

## Solve the inequality

$$
2\sqrt{R^2 - d^2} > \sqrt{3}R
$$

Square both sides:
$$
4(R^2 - d^2) > 3R^2
$$

$$
4R^2 - 4d^2 > 3R^2
$$

$$
R^2 > 4d^2
$$

$$
d < \frac{R}{2}
$$

So the event occurs when the chord is within distance $R/2$ from the center.

---

## Model 1: Random midpoint uniformly in the disk

The chord is determined by choosing its midpoint uniformly in the circle.

Probability is proportional to area:
$$
P = \frac{\pi (R/2)^2}{\pi R^2} = \frac{1}{4}
$$

---

## Model 2: Random radius, perpendicular chord

Choose a random radius and a random point uniformly along it, then draw a perpendicular chord.

Here $d \sim \text{Uniform}[0,R]$

$$
P(d < R/2) = \frac{R/2}{R} = \frac{1}{2}
$$

---

## Model 3: Random endpoints on the circumference

Pick two random points independently on the circle.

This leads to:
$$
P = \frac{1}{3}
$$

---

## Final Answers

$$
\boxed{P = \frac{1}{4}} \quad \text{(random midpoint)}
$$

$$
\boxed{P = \frac{1}{2}} \quad \text{(random radius)}
$$

$$
\boxed{P = \frac{1}{3}} \quad \text{(random endpoints)}
$$

---

## Conclusion

Different mathematically natural definitions of “random chord” produce different probabilities.

This is known as **Bertrand’s Paradox**: the phrase “at random” is ambiguous unless the method of randomization is clearly specified.

 # Q11) CONDITIONAL EXPECTATION

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer’s risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets ${ \omega_1,\omega_2}$ and ${\omega_3,\omega_4}$.

4. Compute $E[X\mid Y]$ explicitly.

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Compute $E[X]$ and compare it with $E[X\mid Y]$.

7. Interpret $E[X\mid Y]$ in practical terms. What does it tell the bank?

8. Let

$$
\mathfrak G=\mathfrak P(\Omega).
$$

Compute $E[X\mid \mathfrak G]$ and compare it with $E[X\mid Y]$.

9. Explain how this example shows that conditional expectation is the “best estimate of $X$ using only the information contained in $Y$.”

# Conditional Expectation: Fraud Detection Model

---

## 1. Probability Space

$$
\Omega = \{\omega_1,\omega_2,\omega_3,\omega_4\}
$$

$$
\mathfrak{F} = \mathfrak{P}(\Omega)
$$

$$
P(\omega_1)=0.50,\quad P(\omega_2)=0.10,\quad P(\omega_3)=0.20,\quad P(\omega_4)=0.20
$$

---

## 2. Sigma-algebra generated by $Y$

Values of $Y$:
- $Y=0$ for $\{\omega_1,\omega_2\}$
- $Y=1$ for $\{\omega_3,\omega_4\}$

Thus:
$$
\sigma(Y) = \{\emptyset,\Omega,\{\omega_1,\omega_2\},\{\omega_3,\omega_4\}\}
$$

---

## 3. Measurability of $E[X|Y]$

$E[X|Y]$ must be $\sigma(Y)$-measurable.

Therefore it must be constant on:
$$
\{\omega_1,\omega_2\}, \quad \{\omega_3,\omega_4\}
$$

---

## 4. Compute $E[X|Y]$

### For $Y=0$:
$$
E[X|Y=0] = \frac{0\cdot0.50 + 1\cdot0.10}{0.50+0.10} = \frac{0.10}{0.60} = \frac{1}{6}
$$

### For $Y=1$:
$$
E[X|Y=1] = \frac{0\cdot0.20 + 1\cdot0.20}{0.20+0.20} = \frac{0.20}{0.40} = \frac{1}{2}
$$

Thus:
$$
E[X|Y](\omega) =
\begin{cases}
\frac{1}{6}, & \omega \in \{\omega_1,\omega_2\} \\
\frac{1}{2}, & \omega \in \{\omega_3,\omega_4\}
\end{cases}
$$

---

## 5. Verification property

For $A = \{\omega_1,\omega_2\}$:

$$
\int_A E[X|Y]\,dP = \frac{1}{6}(0.50+0.10)=0.10
$$

$$
\int_A X\,dP = 0\cdot0.50 + 1\cdot0.10 = 0.10
$$

For $A = \{\omega_3,\omega_4\}$:

$$
\int_A E[X|Y]\,dP = \frac{1}{2}(0.20+0.20)=0.20
$$

$$
\int_A X\,dP = 0\cdot0.20 + 1\cdot0.20 = 0.20
$$

Verified.

---

## 6. Compute $E[X]$

$$
E[X] = 0(0.50)+1(0.10)+0(0.20)+1(0.20)=0.30
$$

Comparison:
- $E[X]=0.30$
- $E[X|Y]=\{1/6, 1/2\}$ (depends on risk group)

---

## 7. Interpretation

$E[X|Y]$ gives the probability of fraud **given the customer’s risk class**:

- Low-risk: $1/6 \approx 0.167$
- High-risk: $1/2 = 0.5$

---

## 8. Conditional on full information

$$
\mathfrak{G} = \mathfrak{P}(\Omega)
$$

Then:
$$
E[X|\mathfrak{G}] = X
$$

Comparison:
- Full information → exact fraud indicator
- Partial info ($Y$) → averaged estimate

---

## 9. Key Insight

$E[X|Y]$ is the best estimate of $X$ using only information in $Y$.

- It minimizes mean squared error
- It uses only observable information (risk class)

---

## Final Insight

- Less information ⇒ more averaging
- More information ⇒ more precise prediction

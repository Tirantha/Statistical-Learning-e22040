# Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.





# 1. Probability Space

Define the probability space as

$$
(\Omega,\mathfrak F,P)
$$

where

## Sample Space

$$
\Omega=\{0,1,2,3\}
$$

## Sigma-algebra

$$
\mathfrak F = 2^\Omega
$$

## Probability Measure

$$
P(\{0\})=0.50
$$

$$
P(\{1\})=0.30
$$

$$
P(\{2\})=0.15
$$

$$
P(\{3\})=0.05
$$

Since

$$
0.50+0.30+0.15+0.05=1
$$

this defines a valid probability measure.

---

# 2. Random Variable

Define

$$
X:\Omega\to\mathbb R
$$

by

$$
X(\omega)=\omega
$$

for all $\omega\in\Omega$.

Thus,

$$
X(0)=0,\quad
X(1)=1,\quad
X(2)=2,\quad
X(3)=3
$$

---

# 3. Probability Mass Function

The probability mass function of $X$ is

$$
p_X(x)=P(X=x)
$$

given by

$$
p_X(x)=
\begin{cases}
0.50, & x=0,\\
0.30, & x=1,\\
0.15, & x=2,\\
0.05, & x=3,\\
0, & \text{otherwise}
\end{cases}
$$

---

# 4. Expectation

$$
E[X]=\sum_x xP(X=x)
$$

$$
=0(0.50)+1(0.30)+2(0.15)+3(0.05)
$$

$$
=0+0.30+0.30+0.15
$$

$$
=0.75
$$

Therefore,

$$
\boxed{E[X]=0.75}
$$

---

# 5. Variance

First compute

$$
E[X^2]
$$

$$
E[X^2]
=0^2(0.50)+1^2(0.30)+2^2(0.15)+3^2(0.05)
$$

$$
=0+0.30+0.60+0.45
$$

$$
=1.35
$$

Now,

$$
\mathrm{Var}(X)=E[X^2]-(E[X])^2
$$

$$
=1.35-(0.75)^2
$$

$$
=1.35-0.5625
$$

$$
=0.7875
$$

Hence,

$$
\boxed{\mathrm{Var}(X)=0.7875}
$$

---

# 6. Required Probabilities

## Probability that at least one failure occurs

$$
P(X\ge1)=P(X=1)+P(X=2)+P(X=3)
$$

$$
=0.30+0.15+0.05
$$

$$
=0.50
$$

Therefore,

$$
\boxed{P(X\ge1)=0.50}
$$

---

## Probability that at least two failures occur

$$
P(X\ge2)=P(X=2)+P(X=3)
$$

$$
=0.15+0.05
$$

$$
=0.20
$$

Therefore,

$$
\boxed{P(X\ge2)=0.20}
$$

---

# 7. Interpretation

The expectation

$$
E[X]=0.75
$$

represents the average number of transformer failures per day.

The variance

$$
\mathrm{Var}(X)=0.7875
$$

measures how much the daily number of failures varies around the mean value.

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.




# 1. Probability Space

Define the probability space as

$$
(\Omega,\mathfrak F,P)
$$

where

## Sample Space

$$
\Omega=[0,1]
$$

since $X$ represents a proportion.

## Sigma-algebra

$$
\mathfrak F=\mathcal B([0,1])
$$

where $\mathcal B([0,1])$ denotes the Borel sigma-algebra on $[0,1]$.

## Probability Measure

For any measurable set $A\subseteq [0,1]$,

$$
P(A)=\int_A f(x)\,dx.
$$

---

# 2. Definition of the Random Variable

Define the random variable

$$
X:\Omega\to\mathbb R
$$

by

$$
X(\omega)=\omega,
\qquad \omega\in[0,1].
$$

Thus, $X$ measures the defective proportion in the batch.

---

# 3. Probability Density Function

The probability density function is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x(1-x)^{17}, & 0<x<1,\\
0, & \text{otherwise}.
\end{cases}
$$

where

$$
B(2,18)=\int_0^1 x^{1}(1-x)^{17}\,dx.
$$

---

# 4. Verification that $f$ is a Valid Density Function

A probability density function must satisfy:

## Non-negativity

For $0<x<1$,

$$
x>0
\qquad \text{and} \qquad
(1-x)^{17}>0.
$$

Hence,

$$
f(x)\ge0.
$$

## Total Integral Equals 1

Compute

$$
\int_{-\infty}^{\infty} f(x)\,dx
=
\int_0^1 \frac{1}{B(2,18)}x(1-x)^{17}\,dx.
$$

Using the definition of the Beta function,

$$
\int_0^1 x(1-x)^{17}\,dx=B(2,18).
$$

Therefore,

$$
\int_{-\infty}^{\infty} f(x)\,dx
=
\frac{1}{B(2,18)}B(2,18)
=1.
$$

Hence, $f$ is a valid probability density function.

---

# 5. Expectation

For a Beta distribution,

$$
E[X]=\frac{\alpha}{\alpha+\beta}.
$$

With $\alpha=2$ and $\beta=18$,

$$
E[X]=\frac{2}{2+18}
=\frac{2}{20}
=0.10.
$$

Therefore,

$$
\boxed{E[X]=0.10}
$$

---

# 6. Variance

For a Beta distribution,

$$
\mathrm{Var}(X)
=
\frac{\alpha\beta}
{(\alpha+\beta)^2(\alpha+\beta+1)}.
$$

Substituting $\alpha=2$ and $\beta=18$,

$$
\mathrm{Var}(X)
=
\frac{2(18)}
{(20)^2(21)}.
$$

$$
=
\frac{36}{8400}
=
\frac{3}{700}.
$$

Thus,

$$
\boxed{\mathrm{Var}(X)=\frac{3}{700}\approx0.004286}
$$

---

# 7. Probability that $X\le0.10$

$$
P(X\le0.10)
=
\int_0^{0.10}
\frac{1}{B(2,18)}
x(1-x)^{17}\,dx.
$$

Therefore,

$$
\boxed{
P(X\le0.10)
=
\frac{1}{B(2,18)}
\int_0^{0.10} x(1-x)^{17}\,dx
}
$$

---

# 8. Probability that $0.05\le X\le0.15$

$$
P(0.05\le X\le0.15)
=
\int_{0.05}^{0.15}
\frac{1}{B(2,18)}
x(1-x)^{17}\,dx.
$$

Hence,

$$
\boxed{
P(0.05\le X\le0.15)
=
\frac{1}{B(2,18)}
\int_{0.05}^{0.15} x(1-x)^{17}\,dx
}
$$

---

# 9. Why the Beta Distribution is Reasonable

The Beta distribution is suitable because:

- it is defined on the interval $[0,1]$,
- proportions and probabilities always lie between $0$ and $1$,
- the shape of the distribution can model many different production behaviors,
- the parameters $\alpha=2$ and $\beta=18$ describe a process where small defective proportions are much more likely than large ones.

---

# 10. Interpretation of Expectation and Variance

The expectation

$$
E[X]=0.10
$$

means that the average defective proportion is approximately $10\%$.

The variance

$$
\mathrm{Var}(X)=\frac{3}{700}
$$

measures how much the defective proportion varies from batch to batch.

A small variance indicates that the production quality is relatively stable over time.

# Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

# PERT Distribution Problem

Given:

$$
a=5,\qquad c=11
$$

and the expected value of the PERT distribution is equal to the most likely value:

$$
E[X]=c.
$$

For a PERT distribution,

$$
E[X]=\frac{a+4c+b}{6}.
$$

Since $E[X]=c$, substitute:

$$
c=\frac{a+4c+b}{6}.
$$

Substitute the given values:

$$
11=\frac{5+4(11)+b}{6}.
$$

Simplify:

$$
11=\frac{5+44+b}{6}
$$

$$
11=\frac{49+b}{6}.
$$

Multiply both sides by $6$:

$$
66=49+b.
$$

Therefore,

$$
b=66-49
$$

$$
b=17.
$$

Hence,

$$
\boxed{b=17}
$$

---

# Variance

Using the PERT approximation,

$$
\mathrm{Var}(X)
=
\frac{(E[X]-a)(b-E[X])}{7}.
$$

Since

$$
E[X]=11,\qquad a=5,\qquad b=17,
$$

substitute:

$$
\mathrm{Var}(X)
=
\frac{(11-5)(17-11)}{7}.
$$

$$
=
\frac{(6)(6)}{7}
$$

$$
=
\frac{36}{7}.
$$

Therefore,

$$
\boxed{
\mathrm{Var}(X)=\frac{36}{7}\approx5.14
}
$$

The standard deviation is

$$
\sigma=\sqrt{\frac{36}{7}}
=\frac{6}{\sqrt7}
\approx2.27.
$$

Thus,

$$
\boxed{\sigma\approx2.27}
$$

days.

---

# Interpretation

Since the expected time equals the most likely time,

$$
E[X]=c=11,
$$

the distribution becomes symmetric around $11$ days.

The minimum and maximum times are equally distant from the mean:

$$
11-5=6,
\qquad
17-11=6.
$$

This symmetry produces balanced uncertainty on both sides of the expected completion time.
